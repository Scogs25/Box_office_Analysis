```python
import pandas as pd
```

```python
Movies_mojo=pd.read_csv('codes/Box_Office_Mojo_Domestic_Box_Office_Chart.csv')
Movies_mojo.head()
```

```python
Movies_imdb=pd.read_csv('zippedData/tmdb.movies.csv')
Movies_imdb.head()
```

```python
Movies_imdb['title_lowercase']=Movies_imdb['original_title'].apply(lambda x:x.lower().replace(' ',''))
```

```python
Movies_mojo['title_lowercase']=Movies_mojo['Title'].apply(lambda x:x.lower().replace(' ',''))
```

```python
df=Movies_mojo.merge(Movies_imdb,how='left',on='title_lowercase')
```

```python
df.drop(axis=1,columns="Unnamed: 0_x",inplace=True)
```
